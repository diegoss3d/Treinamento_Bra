```
# 19. Crie um programa em Python que receba uma string como entrada e imprima "Palíndromo" se a string for um palíndromo (ou seja, se a string lida de trás para frente for igual à string original), ou "Não Palíndromo" caso contrário.
def palindromo(texto):
  if texto == texto[::-1]:
    return "Palíndromo"
  else:
    return "Não Palíndromo"
  
palindromo("cocada")
```

```
# 20. Crie um programa em Python que receba uma lista de números inteiros como entrada e calcule a soma dos quadrados dos números ímpares da lista.
import math 
lista = [2, 5, 4, 2, 3, 6, 5, 4, 2, 3, 8, 5]
soma = 0

for num in lista:
  if num % 2 == 1:
    soma += math.sqrt(num)
    
print(round(soma, 3))
```

```
# 23. Crie um programa em Python que receba uma lista de números e imprima a soma dos números pares e a soma dos números ímpares presentes na lista.
somaPar = 0
somaImpar = 0
lista = [5, 8, 1, 6, 3, 4, 9, 2, 7, 4, 2]
for num in lista:
  if num % 2 == 1:
    somaImpar += num
  else:
    somaPar += num
print(f"A soma dos pares é {somaPar}, a soma dos ímpares é {somaImpar}")
```

```
# 25. Crie um programa em Python que leia um número do usuário e verifique se ele é um número perfeito. Um número perfeito é aquele que é igual à soma de seus divisores próprios (ou seja, a soma de seus fatores inteiros positivos, excluindo ele próprio).
numero = int(input("Digite um numero para saber se ele é um numero perfeito:")) # 28  2, 3
soma = 0
for num in range(1, numero):
  if numero % num == 0:
    soma+= num
if soma == numero:
  print(f"O número {numero} é um numero perfeito")
else:
  print(f"O número {numero} não é um numero perfeito")
```
